```python
#Task 1
import random

class Creature:

    def __init__(self, Name, MaxHP=10):
        self.name=Name
        self.maxHP=MaxHP
        self.HP=MaxHP
        self.abilities={"Attack":1, "Defense":5, "Speed":5}
    
    def check_life(self,test):
        if (self.HP<=0):
            self.HP=0
            if test==True:
                print(self.name, "died.")
            return self.HP ##which is 0
        else:
            return self.HP
    
    def attack(self, target):
        self.roll=random.randint(1,20)
        print(self.name,"attacks", target.name+".")
        if(self.roll<(target.abilities["Defense"]+target.abilities["Speed"])):##defense and speed values
            ##attack not successful
            print("Attack missed...")
        else:
            attack_val=self.abilities["Attack"]+random.randint(1,4)
            target.HP=target.HP-attack_val
            print("Attack was successful, dealing", abs(attack_val),"damage points!")

    def turn(self, round_num, target,p_rint):
        #print is used to print the rounds in error checking
        if p_rint==True:
            print("Round ",round_num,":")
        self.attack(target)
        if(target.check_life(True)==0):
            ##return false if target has died
            return False
        else:
            return True
```

```python
#Task 2
class Fighter(Creature):
    shield_is_up=False##flag to see if shield is up or not

    def __init__(self, Name, MaxHP=50):
        self.name=Name
        self.maxHP=MaxHP
        self.HP=MaxHP
        self.abilities={"Attack":5, "Defense":10, "Speed":3}

    def shield_up(self):
        if self.shield_is_up==False:
            self.abilities["Attack"]=self.abilities["Attack"]-5
            self.abilities["Defense"]=self.abilities["Defense"]+5
            print("Shield was put up")
            self.shield_is_up=True
        else:
            print("Shield is already up")##no points are added

    def shield_down(self):
        if self.shield_is_up==True:
            self.abilities["Attack"]=self.abilities["Attack"]+5
            self.abilities["Defense"]=self.abilities["Defense"]-5
            print("Shield was put down")
            self.shield_is_up=False
        else:
            print("Shield is already down")##no points are added

    def turn(self,round_num, target,p_rint):
        if p_rint==True:
            print("Round ",round_num,":")
        if round_num%4==1:
            self.attack(target)
            self.shield_up()
        elif(round_num%4==2 or round_num%4==3):
            self.attack(target)
        else:
            self.shield_down()
            self.attack(target)

        if(target.check_life(True)==0):
            ##return false if target has died
            return False
        else:
            return True
```

```python
#Task3
class Archer(Creature):
    abilities_change=False ##flag variable
    abil_attack=False##flag

    def __init__(self, Name, MaxHP=30):
        self.name=Name
        self.maxHP=MaxHP
        self.HP=MaxHP
        self.abilities={"Attack":7, "Defense":8, "Speed":8}
    
    def sneak_attack(self, target):
        a=random.randint(1,20)
        b=random.randint(1,20)
        if a>b:
            self.abilities["Attack"]=a
        else:
            self.abilities["Attack"]=b
        
        if self.abilities["Speed"]>target.abilities["Speed"]:
            print(self.name+"'s attack has risen.")
            print(self.name+"'s defense has reduced.")
            self.abilities["Attack"]+=self.abilities["Speed"]-target.abilities["Speed"]
        
        if self.abilities_change==False:
            self.abilities["Attack"]+=3
            self.abilities["Defense"]-=3
            self.abilities_change=True

        self.roll=random.randint(1,8)+self.abilities["Attack"]
        print(self.name,"sneak attacks", target.name+"...")
        if(self.roll<(target.abilities["Defense"]+target.abilities["Speed"])):##defense and speed values
            ##attack not successful
            print("Attack missed...")
        else:
            attack_val=self.abilities["Attack"]+random.randint(1,4)
            target.HP=target.HP-attack_val
            print("Sneak attack was successful, dealing", abs(attack_val),"damage points!")
        
    def attack(self,target):
        if self.abil_attack==False:
            self.abilities["Attack"]=7
            self.abilities["Defense"]=8
            self.abil_attack=True#prevents further calling
        
        Creature.attack(self,target)##call attack from creature
    
    def turn(self, round_num, target, p_rint):
        if p_rint==True:
            print("Round ",round_num,":")
        if (round_num%4==1 or round_num%4==0):
            self.attack(target)
        else:
            self.sneak_attack(target)

        if(target.check_life(True)==0):
            ##return false if target has died
            return False
        else:
            return True
```

```python
#Task1
class Goblin(Creature):

    def __init__(self, Name, MaxHP=15):
        self.name=Name
        self.maxHP=MaxHP
        self.HP=MaxHP
        self.abilities={"Attack":4, "Defense":6, "Speed":6}


class Orc(Creature):
    modified=False
    a_mod=False
    n=5

    def __init__(self, Name, MaxHP=50):
        self.name=Name
        self.maxHP=MaxHP
        self.HP=MaxHP
        self.abilities={"Attack":10, "Defense":6, "Speed":2}
    
    def heavy_attack(self, target):
        print(self.name+" is in rage.")
        if self.modified==False:
            self.abilities["Attack"]+=5
            self.abilities["Defense"]-=3
            self.modified=True
            self.a_mod=False

        Creature.attack(self,target)##call attack from creature
    
    def attack(self, target):
        if self.a_mod==False:
            self.abilities["Attack"]=10
            self.abilities["Defense"]=6
            self.a_mod=True
            self.modified=False##resetting values

        Creature.attack(self,target)##call attack from creature

    def turn(self, round_num, target, p_rint):
        if p_rint==True:
            print("Round ",round_num,":")
        if (round_num%4==0):
            self.heavy_attack(target)
        elif (round_num%self.n==0):##self.n required for correct orc cooling
            print("Orc cooled down")
            self.attack(target)
            self.n=self.n+4
        else:
            self.attack(target)

        if(target.check_life(True)==0):
            ##return false if target has died
            return False
        else:
            return True
```

```python
#Task1

class Wizard(Creature):
    def __init__(self, Name, MaxHP=20):
        self.name=Name
        self.maxHP=MaxHP
        self.HP=MaxHP
        self.abilities={"Attack":3, "Defense":5, "Speed":5, "Arcana":10}
        self.max_mana=100
        self.mana=100##needs to be in the range 0-100
#Task2
    def attack(self, target):
        if (self.mana+20>100):
            self.mana=100
            print("Mana is now full")
        else:
            self.mana+=20
            print("Mana: +20!")
        Creature.attack(self,target)

    def recharge(self):
        if (self.mana+30>100):
            self.mana=100
            print("Mana is now full")
        else:
            self.mana+=30
            print(self.name, "channels magical energy...")
            print("Mana: +30!")

    def fire_bolt(self, target):
        self.roll=random.randint(1,20)
        print(self.name,"fires a fire bolt at",target.name+"...")
        if(self.roll<(target.abilities["Defense"]+target.abilities["Speed"])):##defense and speed values
            ##attack not successful
            print("Attack missed...")
        else:
            attack_val=self.abilities["Attack"]+random.randint(1,4)+ self.abilities["Arcana"]//2
            if (attack_val>self.abilities["Arcana"]):
                attack_val=self.abilities["Arcana"]
            target.HP=target.HP-attack_val
            print("Fire bolt was successful, dealing", abs(attack_val),"fire damage points!")
            if (self.mana+10>100):
                self.mana=100
                print("Mana is now full")
            else:
                self.mana+=10
                print("Mana: +10!")

    def heal(self,target):
        if (self.mana<20):
            print("Heal has failed due to insufficient mana")
        else:
            self.mana-=20
            print("Mana: -20")
            self.heal_amount=random.randint(0,8)+self.abilities["Arcana"]//2

            if (target.HP+self.heal_amount>target.maxHP):
                target.HP=target.maxHP
                print(self.name+" heals "+target.name+" to full HP!")
            else:
                target.HP+=self.heal_amount
                self.heal_amount=str(self.heal_amount)
                print(self.name+" heals "+target.name+" for "+self.heal_amount+" HP!")

    def mass_heal(self, allies):
        if (self.mana<30):
            print("Mass heal has failed due to insufficient mana")
        else:
            self.mana-=30
            print("Mana: -30")
            ##loop here for each ally
            for i in range(len(allies)):
                self.heal_amount=random.randint(0,10)+self.abilities["Arcana"]
                allies[i].HP+=self.heal_amount
                if (allies[i].HP+self.heal_amount>allies[i].maxHP):
                    allies[i].HP=allies[i].maxHP
                    print(self.name+" heals "+allies[i].name+" to full HP!")
                else:
                    allies[i].HP+=self.heal_amount
                    self.heal_amount=str(self.heal_amount)
                    print(self.name+" heals "+allies[i].name+" for "+self.heal_amount+" HP!")


    def fire_storm(self, enemies):
        if (self.mana<50):
            print("Fire storm has failed due to insufficient mana")
        else:
            self.mana-=50
            print("Mana: -50")
            ####loop for enemies here
            for i in range(len(enemies)):
                self.number=random.randint(1,20)+enemies[i].abilities["Speed"]
                self.storm=random.randint(5,20)+self.abilities["Arcana"]
                if(self.number>=self.abilities["Arcana"]):
                    self.storm=self.storm//2
                enemies[i].HP=enemies[i].HP-self.storm
                print("Fire Storm deals", self.storm,"fire damage to",enemies[i].name+"!")
```
